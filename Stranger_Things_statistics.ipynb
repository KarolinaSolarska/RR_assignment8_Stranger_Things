# Stranger Things

**Stranger Things** is an American science fiction horror drama television series created by the Duffer Brothers for **Netflix**. Produced by Monkey Massacre Productions and 21 Laps Entertainment, the first season was released on Netflix on **July 15, 2016**. The second and third seasons followed in October 2017 and July 2019 respectively, and the fourth season was released in two parts in May and July 2022. In February 2022, Stranger Things was renewed for a fifth and final season, which is expected to be released in 2025.

```python
import pandas as pd

# Inputting the data from Wikipedia
data = {
    "Season": [1, 2, 3, 4, 4, 5],
    "Episodes": [8, 9, 8, 7, 2, 8],
    "Originally released": ["2016", "2017", "2019", "2022", "2022", "2025"]
}

# Creating the dataframe
df_stranger_things = pd.DataFrame(data)
df_stranger_things
```

```python
import matplotlib.pyplot as plt

# Converting the 'Originally released' to datetime
df_stranger_things['Originally released'] = pd.to_datetime(df_stranger_things['Originally released'], errors='coerce')

# Basic statistics for the number of episodes
episode_stats = df_stranger_things['Episodes'].describe()

# Printing basic statistics
print(episode_stats)
```

**Number of Episodes per Season** 

This bar chart shows the number of episodes in each season of "Stranger Things". The season numbers are along the x-axis, and the number of episodes is on the y-axis.

```python
# Plotting the number of episodes per season
plt.figure(figsize=(10, 6))
plt.bar(df_stranger_things['Season'].astype(str), df_stranger_things['Episodes'], color='skyblue')
plt.xlabel('Season')
plt.ylabel('Number of Episodes')
plt.title('Number of Episodes per Season of Stranger Things')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Release Timeline** 

This line graph shows the release dates for each season (except for season 5, which doesn't have a specific release date yet) along a timeline.

```python
# Plotting the release timeline
df_plot = df_stranger_things.dropna(subset=['Originally released'])

plt.figure(figsize=(10, 6))
plt.plot_date(df_plot['Originally released'], df_plot['Season'].astype(str), 'o-')
plt.xlabel('Date')
plt.ylabel('Season')
plt.title('Release Timeline of Stranger Things')
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the bar chart illustrating the number of episodes per season of "Stranger Things," we observe that seasons 1 and 3 each have 8 episodes, which seems to be the standard for the series, given that this number appears most frequently. Season 2 has the highest number of episodes at 9.

The release timeline graph reveals the pattern of release dates for each season. Seasons 1 to 2 were released approximately one year apart, with season 3 being released 2 years after. There's a noticeable longer period of time between sezon 3 and 4, more or less 3 years. The expected date of publication of the 5th part is 2025.
